#### PE trimming

```python
!trimmomatic PE -phred33 /content/01_raw_data/SRR1799908_1.fastq \
/content/01_raw_data/SRR1799908_2.fastq SRR1799908_output_R1_paired_trimmed.fastq \
 SRR1799908_output_R1_unpaired_trimmed.fastq \
 SRR1799908_output_R2_paired_trimmed.fastq \
 SRR1799908_output_R2_unpaired_trimmed.fastq \
 ILLUMINACLIP:/content/03_trimming/trimmomatic/adapters/NexteraPE-PE.fa:2:40:15 SLIDINGWINDOW:4:20 MINLEN:25
```

Explanation by component:

trimmomatic
    Calls the Trimmomatic program.

    PE
    Indicates paired-end read data (forward and reverse reads).

    -phred33
    Specifies that the quality scores are encoded using Phred+33 (common for Illumina reads).

Input Files (2):

    /content/01_raw_data/SRR17864445_1.fastq
    Forward reads.

    /content/01_raw_data/SRR17864445_2.fastq
    Reverse reads.

Output Files (4):

    SRR17864445_output_R1_paired_trimmed.fastq
    Forward reads that passed trimming (high quality, adapter-free).

    SRR17864445_output_R1_unpaired_trimmed.fastq
    Forward reads that lost their pair due to trimming (e.g., one read was removed).

    SRR17864445_output_R2_paired_trimmed.fastq
    Reverse reads that passed trimming.

    SRR17864445_output_R2_unpaired_trimmed.fastq
    Reverse reads that lost their pair.

Trimming Options:

    ILLUMINACLIP:/content/03_trimming/trimmomatic/adapters/NexteraPE-PE.fa:2:40:15
    Removes adapter sequences:

        NexteraPE-PE.fa is the adapter file.

        2: Maximum mismatches allowed in seed.

        40: Palindrome clip threshold (used for paired-end).

        15: Simple clip threshold.

    SLIDINGWINDOW:4:20
    Performs quality trimming:

        Scans with a 4-base sliding window.

        Trims when average quality in the window falls below 20.

    MINLEN:25
    Discards reads that are shorter than 25 bases after trimming.

**BCR::ABL fusion gene**
[The Philadelphia (Ph) chromosome](https://www.cancer.gov/publications/dictionaries/cancer-terms/def/bcr-abl-fusion-gene) is an abnormal chromosome that is made when pieces of chromosomes 9 and 22 break off and trade places.
```python
import os

# Directory containing VCF files
vcf_dir = "/content/05_VC/isec_dir"
output_main_dir = "/content/06_snpEff_reports"

# Ensure main output directory exists
os.makedirs(output_main_dir, exist_ok=True)

# Annotate each VCF file
vcf_files = [f for f in os.listdir(vcf_dir) if f.endswith(".vcf")]

for vcf_file in vcf_files:
    input_vcf = os.path.join(vcf_dir, vcf_file)

    # Create a directory for the current VCF file
    output_dir = os.path.join(output_main_dir, os.path.splitext(vcf_file)[0])
    os.makedirs(output_dir, exist_ok=True)

    # Annotate VCF file using snpEff
    annotated_vcf = os.path.join(output_dir, f"annotated_{vcf_file}")
    command = f"java -Xmx4g -jar /content/snpEff/snpEff.jar eff -cancer -v GRCh37.75 {input_vcf} > {annotated_vcf}"
    os.system(command)

    # Generate statistics report for the annotated VCF file
    stats_file = os.path.join(output_dir, f"stats_{vcf_file}.html")
    command_stats = f"java -jar /content/snpEff/SnpSift.jar stats {annotated_vcf} > {stats_file}"
    os.system(command_stats)
```
